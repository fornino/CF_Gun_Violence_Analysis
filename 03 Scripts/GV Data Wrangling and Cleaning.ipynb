# 04. Deriving New Columns

```python
# Following columns have unusable data in current form: gun stolen, incident_characteristics, participant_age, participant age_group, participant_gender, participant_type
df_cols = df_2[['gun_stolen', 'incident_characteristics', 'participant_age', 'participant_age_group', 'participant_gender', 'participant_type']]
df_cols.head()
```

## gun_stolen flag column

```python
df_2['gun_stolen'].value_counts()
```

```python
stolen_gun_series = pd.Series(df_2['gun_stolen'])
len(stolen_gun_series)
```

```python
# For loop to create new stolen_gun_involved column
stolen_gun = []
for x in range(0, len(stolen_gun_series)):
    if ('Stolen' in (stolen_gun_series.iloc[x])):
        stolen_gun.append('Stolen')
    elif ('Unknown' in (stolen_gun_series.iloc[x])):
        stolen_gun.append('Unknown')
    else:
        stolen_gun.append('Not Stolen')
df_2['stolen_gun_involved'] = stolen_gun
```

```python
df_2['stolen_gun_involved'].value_counts()
```

## incident_characteristics flag columns

```python
df_2['incident_characteristics'].value_counts()
```

```python
incidents = pd.Series(df_2['incident_characteristics'])
len(incidents)
```

```python
possession_by_felon = []
for x in range(0, len(incidents)):
    if ('Possession of gun by felon' in (incidents.iloc[x])):
        possession_by_felon.append('True')
    else:
        possession_by_felon.append('False')
df_2['possession_by_felon'] = possession_by_felon
df_2['possession_by_felon'].value_counts()
```

```python
shooting_incident = []
for x in range(0, len(incidents)):
    if ('Non-Shooting Incident' in (incidents.iloc[x])):
        shooting_incident.append('False')
    else:
        shooting_incident.append('True')
df_2['shooting_incident'] = shooting_incident
df_2['shooting_incident'].value_counts()
```

```python
df_2 = df_2[df_2['shooting_incident'] == 'True']
df_2.shape
```

```python
df_exclusion = df_2[(df_2['possession_by_felon'] == 'True') & (df_2['shooting_incident'] == 'True')]
df_exclusion.head()
```

```python
df_exclusion['incident_characteristics'].value_counts()
```

```python
incidents = pd.Series(df_exclusion['incident_characteristics'])
len(incidents)
```

```python
# Drive-by flag
drive_by = []
for x in range(0, len(incidents)):
    if ('Drive-by' in (incidents.iloc[x])):
        drive_by.append('True')
    else:
        drive_by.append('False')
df_exclusion['drive_by_flag'] = drive_by
df_exclusion['drive_by_flag'].value_counts()
```

```python
# Domestic Violence flag
domestic_violence = []
for x in range(0, len(incidents)):
    if ('Domestic Violence' in (incidents.iloc[x])):
        domestic_violence.append('True')
    else:
        domestic_violence.append('False')
df_exclusion['domestic_violence_flag'] = domestic_violence
df_exclusion['domestic_violence_flag'].value_counts()
```

```python
# Armed Robbery flag
armed_robbery = []
for x in range(0, len(incidents)):
    if ('Armed robbery' in (incidents.iloc[x])):
        armed_robbery.append('True')
    else:
        armed_robbery.append('False')
df_exclusion['armed_robbery_flag'] = armed_robbery
df_exclusion['armed_robbery_flag'].value_counts()
```

```python
# Gang Involvement
gang_involvement = []
for x in range(0, len(incidents)):
    if ('Gang involvement' in (incidents.iloc[x])):
        gang_involvement.append('True')
    else:
        gang_involvement.append('False')
df_exclusion['gang_involvement_flag'] = gang_involvement
df_exclusion['gang_involvement_flag'].value_counts()
```

```python
# Drug involvement flag
drug_involvement = []
for x in range(0, len(incidents)):
    if ('Drug involvement' in (incidents.iloc[x])):
        drug_involvement.append('True')
    else:
        drug_involvement.append('False')
df_exclusion['drug_involvement_flag'] = drug_involvement
df_exclusion['drug_involvement_flag'].value_counts()
```

```python
# Accidental Shooting flag
accidental = []
for x in range(0, len(incidents)):
    if ('Accidental' in (incidents.iloc[x])):
        accidental.append('True')
    else:
        accidental.append('False')
df_exclusion['accidental_flag'] = accidental
df_exclusion['accidental_flag'].value_counts()
```

```python
# Self Inflicted Shooting flag
self_inflicted= []
for x in range(0, len(incidents)):
    if ('Self-Inflicted' in (incidents.iloc[x])):
        self_inflicted.append('True')
    else:
        self_inflicted.append('False')
df_exclusion['self_inflicted_flag'] = self_inflicted
df_exclusion['self_inflicted_flag'].value_counts()
```

```python
# Mass Shooting flag
mass_shooting = []
for x in range(0, len(incidents)):
    if ('Mass Shooting' in (incidents.iloc[x])):
        mass_shooting.append('True')
    else:
        mass_shooting.append('False')
df_exclusion['mass_shooting_flag'] = mass_shooting
df_exclusion['mass_shooting_flag'].value_counts()
```

```python
df_exclusion.loc[df_exclusion['n_injured'] > 0, 'injuries'] = 'True'
df_exclusion.loc[df_exclusion['n_killed'] > 0, 'deaths'] = 'True'
df_exclusion.loc[(df_exclusion['n_injured'] == 0) & (df_exclusion['n_killed'] == 0), 'no_injuries_or_deaths'] = 'True'
```

```python
df_exclusion['injuries'].fillna(value = 'False', inplace=True)
df_exclusion['deaths'].fillna(value = 'False', inplace=True)
df_exclusion['no_injuries_or_deaths'].fillna(value = 'False', inplace=True)
```

```python
# Child involved flag
child_involved = []
for x in range(0, len(ages)):
    if ('Child 0-11' in (ages.iloc[x])):
        child_involved.append('True')
    else:
        child_involved.append('False')
df_exclusion['child_participant'] = child_involved
df_exclusion['child_participant'].value_counts()
```

```python
# Teen involved flag
teen_involved = []
for x in range(0, len(ages)):
    if ('Teen 12-17' in (ages.iloc[x])):
        teen_involved.append('True')
    else:
        teen_involved.append('False')
df_exclusion['teen_participant'] = teen_involved
df_exclusion['teen_participant'].value_counts()
```

```python
# Adult involved flag
adult_involved = []
for x in range(0, len(ages)):
    if ('Adult 18+' in (ages.iloc[x])):
        adult_involved.append('True')
    else:
        adult_involved.append('False')
df_exclusion['adult_participant'] = adult_involved
df_exclusion['adult_participant'].value_counts()
```

## Participant gender flags

```python
gender = pd.Series(df_exclusion['participant_gender'])
len(gender)
```

```python
# Gender flag: all female
all_female = []
for x in range(0, len(gender)):
    if ('Male' not in (gender.iloc[x])):
        all_female.append('True')
    else:
        all_female.append('False')
df_exclusion['all_female_participants'] = all_female
df_exclusion['all_female_participants'].value_counts()
```

```python
# Gender flag: all male
all_male = []
for x in range(0, len(gender)):
    if ('Female' not in (gender.iloc[x])):
        all_male.append('True')
    else:
        all_male.append('False')
df_exclusion['all_male_participants'] = all_male
df_exclusion['all_male_participants'].value_counts()
```

```python
df_exclusion.loc[(df_exclusion['all_female_participants'] == 'False') & (df_exclusion['all_male_participants'] == 'False'), 'mixed_gender_participants'] = 'True'
df_exclusion['mixed_gender_participants'].fillna('False', inplace = True)
df_exclusion['mixed_gender_participants'].value_counts()
```

## Participant type flag

```python
p_type = pd.Series(df_exclusion['participant_type'])
len(p_type)
```

```python
# Participant type: all victims
all_victims = []
for x in range(0, len(p_type)):
    if ('Subject-Suspect' not in (p_type.iloc[x])):
        all_victims.append('True')
    else:
        all_victims.append('False')
df_exclusion['all_victims'] = all_victims
df_exclusion['all_victims'].value_counts()
```

```python
# Participant type: all suspects
all_suspects = []
for x in range(0, len(p_type)):
    if ('Victim' not in (p_type.iloc[x])):
        all_suspects.append('True')
    else:
        all_suspects.append('False')
df_exclusion['all_suspects'] = all_suspects
df_exclusion['all_suspects'].value_counts()
```

```python
df_exclusion.loc[(df_exclusion['all_victims'] == 'False') & (df_exclusion['all_suspects'] == 'False'), 'suspects_and_victims'] = 'True'
df_exclusion['suspects_and_victims'].fillna('False', inplace = True)
df_exclusion['suspects_and_victims'].value_counts()
```

## Creating n_participants Column

```python
import re
def extractmax(str1):
    nums = re.findall("[0-9]+", str1)
    return (max(map(int, nums)))
```

```python
# Finding max number in gender column to derive n_participants
num_participants = []
for x in range(0, len(gender)):
    num_participants.append(extractmax(gender.iloc[x]))
    
df_exclusion['n_participants'] = num_participants
df_exclusion['n_participants'].value_counts()
```
